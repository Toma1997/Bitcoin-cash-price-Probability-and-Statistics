```python
# uvodjenje potrebnih biblioteka za ucitavanje i manipulisanje skupom podataka
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import locale
from locale import atof
%matplotlib inline

# ucitavanje bitcoin skupa podataka i prikaz prvih 5 redova
data_por = pd.read_csv("bitcoin_cash_price.csv") 
# Volume je novcani iznos dnevne trgovine valutom (prodaje, kupovine)
# Market Cap je trenutna cena * broj bitcoin-ova - to je trenutna ukupna vrednost svih bitcoin-a
data_por
```

```python
atributi = ["Volume", "Market Cap"] # izdvajamo 2 kolone koje treba da formatiramo u brojeve bez zareza
data_por["Market Cap"] = data_por["Market Cap"].replace("-", "0") # zamenjujemo - sa 0 da bi mogli u broj da pretvorimo

data_por['Date'] = pd.to_datetime(data_por['Date'])  #Promena tipa datuma.

# prevodjenje stringa sa zarezom u int64 tip
locale.setlocale(locale.LC_ALL, '')
data_por[atributi] = data_por[atributi].applymap(atof)
data_por[atributi] = data_por[atributi].astype(np.int64)
data_por.dtypes # da proverimo tipove podataka
```

## 1) Estimacija uslovnih i bezuslovnih verovatnoća određenih događaja

```python
# prvo cemo racunati bezuslovnu verovatnocu za 24h kolicinu trgovanja 
data_temp1 = (data_por["Volume"]<500000000).value_counts()[True] # koliko dana je dnevna trgovina manja od 500 miliona dolara
data_temp2 = (data_por["Volume"]>1000000000).value_counts()[True] # koliko dana je dnevna trgovina veca od milijardu dolara
data_temp3 = (len(data_por) - (data_temp1 + data_temp2)) # koliko dana je dnevna trgovina izmedju 500 miliona i milijardu dolara
data_temp = pd.Series({'0':data_temp1, '1':data_temp3, '2':data_temp2}) # napravimo vremensku seriju sa ovim podacima
data_temp
```

```python
P_dnevniPromet = pd.DataFrame(data_temp/data_temp.sum()) # svaki element se podeli sa ukupnim brojem elemenata
P_dnevniPromet.index = ["Ispod 500 miliona dolara","Izmedju 500 miliona i milijardu dolara", "Iznad milijardu dolara"]
P_dnevniPromet.columns = ["Verovatnoca"]
P_dnevniPromet.columns.name = "Dnevna kolicina trgovanja"
print(P_dnevniPromet)
P_dnevniPromet.plot.bar(figsize=(15,9),fontsize=14)
plt.xlabel("Dnevna kolicina trgovanja",fontsize=18)
plt.ylabel("Verovatnoca",fontsize=18)
plt.show()
```

```python
# bezuslovna verovatnoca da cena raste, odnosno pada
# koliko dana je cena rasla na kraju dana u odnosu na pocetak dana
data_temp = (data_por["Close"] > data_por["Open"]).value_counts() 
data_temp
```

```python
P_monotonostCene = pd.DataFrame(data_temp/data_temp.sum())
P_monotonostCene.index = ["Opada cena","Raste cena"]
P_monotonostCene.columns = ["Verovatnoca"]
P_monotonostCene.columns.name = "Dnevna monotonost cene"
print(P_monotonostCene)
P_monotonostCene.plot.bar(figsize=(10,6),fontsize=16)
plt.xlabel("Dnevna monotonost cene",fontsize=18)
plt.ylabel("Verovatnoca",fontsize=18)
plt.show()
```

```python
# Prikazacemo kretanje vrednosti trzista bitcoina u toku celog perioda iz skupa podataka
#Kreiranje serije sa vrednostima iz kolone "Market Cap"
marketCap= pd.Series(data=data_por['Market Cap'].values,index=data_por.Date)
plt.figure(figsize=(16, 8))
plt.ylabel("Vrednost trzista u desetinama milijardi dolara",fontsize=16)
plt.plot(marketCap, c='b') #Crtanje grafika kretanja vrednosti celog trzista bitcoin-a
```

```python
# Sada cemo estimirati uslovnu verovatnocu
# koliko je verovatnoca da cena raste pod uslovom da je dnevna trgovina manja od 500 miliona
data_temp4 = data_por.loc[data_por['Volume']<500000000]
data_temp5 = (data_temp4["Close"] > data_temp4["Open"]).value_counts() # kada cena raste
data_temp5
```

```python
P_mon_pod_usl_dnevne_trgovine= pd.DataFrame((data_temp5/data_temp4.shape[0]))
P_mon_pod_usl_dnevne_trgovine.index = ["Opada cena","Raste cena"]
P_mon_pod_usl_dnevne_trgovine.columns = ["Verovatnoca"]
print("Verovatnoca da cena raste ili opada pod uslovom da je dnevna trgovina manja od 500 miliona:")
print(P_mon_pod_usl_dnevne_trgovine)
P_mon_pod_usl_dnevne_trgovine.columns.name="Dnevna monotonost cene"
P_mon_pod_usl_dnevne_trgovine.plot(kind='bar',figsize=(12,9),fontsize=16)
plt.xlabel("Uslovna dnevna monotonost cene",fontsize=18)
plt.ylabel("Verovatnoca",fontsize=18)
plt.show()
```

Možemo da rezonujemo i na obrnut način: Hoćemo da predvidimo verovatnoću da je dnevna trgovina ispod odredjene sume pod uslovom monotonosti cene. Za računanje ovoga mozemo koristiti Bajesovo pravilo: 

$$P(\text{dnevna trgovina}\ |\  \text{monotonost cene})=\frac{P(\text{monotonost cene}\ |\ \text{dnevna trgovina})P(\text{dnevna trgovina})}{P(\text{monotonost cene})}$$

```python
P_dnevna_trgovina_pod_usl_mon = P_mon_pod_usl_dnevne_trgovine * P_dnevniPromet.loc["Ispod 500 miliona dolara"] / P_monotonostCene
print("Verovatnoca dnevne trgovine pod uslovom da je data verovatnoca monotonosti cene:")
print(P_dnevna_trgovina_pod_usl_mon)
pd.DataFrame(P_dnevna_trgovina_pod_usl_mon).plot.bar(figsize=(12,9),fontsize=18).legend(loc="best")
plt.xlabel("Dnevna monotonost cene",fontsize=18)
plt.ylabel("Verovatnoca",fontsize=18)
plt.show()
```

## 4) Estimacija funkcije gustine raspodele pomoću histograma i testiranje hipoteza o raspodeli korišćenjem Kolmogorov-Smirnov i/ili hi-kvadrat testa

```python
# prvo cemo uraditi estimaciju funkcije gustine raspodele za kolonu Close - cena na kraju dana
data_close = data_por.loc[:, ["Close"]]
data_close = data_close.astype(np.int64)
plt.style.use('seaborn-talk')
hist = data_close.hist(normed=True,bins=100)
plt.show()
```

```python
# Normalna raspodela je izmedju ostalog vazna jer se jako cesto dobija kao aproskimirajuca raspodela slucajnih promenljivih 
# koje predstavljaju sume velikog broja slucajnih promeljivih sa proizvoljnim raspodelama (centralna granicna teorema)
# Probacemo aproksimaciju Gausovom raspodelom, i Gausovom mesavinom sa adekvatnim srednjim vrednostima i varijansama
from scipy.stats import norm

# kreiramo histogram
hist = data_close.hist(normed=True,bins=60)
korak = 0.1
x = np.arange(0, 4000, korak)


m1=400
sig1=195

m2=1450
sig2=310

m3=2400
sig3=50

m4=2600
sig4=320

m5=3300
sig5=640

# Pravimo niz verovatnoca funkcije gustine Normalne (Gausove) raspodele
P_norm_mix5 = np.array(norm.pdf(x, m1, sig1)+0.55*norm.pdf(x, m2, sig2)+
                       0.03*norm.pdf(x, m3, sig3)+0.2*norm.pdf(x, m4, sig4)+0.2*norm.pdf(x, m5, sig5))
fig = plt.gcf()
fig.set_size_inches(16,8)
plt.plot(x, P_norm_mix5, 'b', linewidth=3.0)
plt.show()
```

```python
# Na osnovu gornje aproksimacije odredicemo verovatnocu da je cena na kraju dana izmedju 1000 i 2000 dolara

# Rucno priblizno integraljenje:
PP=np.array([korak*z for x,z in zip(x,P_norm_mix5) if (x>1000 and x<2000)])
P_trazeno=PP.sum(axis=0)
print(P_trazeno)
```

```python
# Sada cemo da radimo testiranje hipoteza o raspodeli korišćenjem Kolmogorov-Smirnov testa
from scipy import stats

# testiracemo da li je ovo normalna raspodele:

velicina_uzorka=100

stats.kstest(data_close.sample(velicina_uzorka).dropna().values.T, cdf='norm',args=(data_close.dropna().values.mean(), data_close.dropna().values.std()))  
# u args stavljamo ocene matematickog ocekivanja i standardne devijacije
```

p vrednost je izuzetno mala sto znaci da cemo odbaciti hipotezu da je u pitanju normalna raspodela za kolonu Close

```python
estimatori_srednje_vrednosti = []       

broj_estimacija_srednjih_vrednosti=50   # ovo je broj estimacija, svaki je baziran na uzorku velicine 'velicina_uzorka'
for x in range(broj_estimacija_srednjih_vrednosti):
    
    uzorak=data_close.sample(velicina_uzorka)
    estimatori_srednje_vrednosti.append( uzorak.dropna().values.mean() )
    
# plotujemo estimaciju gustine verovatnoce raspodele srednjih vrednosti (pomocu histograma) - vidimo da je priblizno normalna!  
pd.DataFrame(estimatori_srednje_vrednosti).plot(kind="density")   

# histogram srednjih vrednosti:
hist = pd.DataFrame(estimatori_srednje_vrednosti).hist(normed=True,bins=30) 
plt.show()
```

```python
# testiracemo da li estimatori imaju normalnu raspodelu:
stats.kstest(estimatori_srednje_vrednosti, cdf='norm',args=(data_close.dropna().values.mean(), data_close.dropna().values.std()/velicina_uzorka**0.5))
```

Za estimatore je veca p vrednost, i njihova raspodela naginje ka normalnoj ali i dalje nije zadovoljavajuca da bi smo rekli da je skroz normalna sto se moze videti i numericki (p vrednost) i po graficima iznad.

```python
# testiranje hipoteze hi-kvadrat testom za raspodelu estimatora
# Hi-kvadrat testom ukratko ispitujemo da li postoji znacajnija razlika izmedju prave raspodele i estimirane raspodele
stats.chisquare(estimatori_srednje_vrednosti) # pretpostavlja se uniformna raspodela posto f_exp nije specificirano
```

Vidimo da je p vrednost izuzetno mala sto znaci da estimatori nisu uopste u uniformnoj raspodeli

## 5) Regresija (Linearna, Visestruka i Nelinearna)

```python
# Prvo primenjujemo Linearnu Regresiju
import statsmodels.formula.api as sm

# trazimo linearnu funkciju zavisnosti izmedju promenljivih High - najvise cene u toku dana i Close - cene na kraju dana
dataReg = data_por.loc[:, ["High", "Low"]]
dataReg.plot('High', 'Low', style='o')
plt.xlim([0, 4500])
plt.ylim([0, 4500])
plt.show()
```

```python
# fit() funkcijom odredjujemo parametre k i n za nasu linearnu funkciju f(x) = k*x + n uz pomoc analiziranja vektora Close i High
Rez1= sm.ols('Low ~ High', dataReg).fit() # ukljucujemo svih 213 tacaka u nas model
Rez2 = sm.ols('Low ~ High', dataReg[:-100]).fit()  # bez poslednjih 100 tacaka
dataReg.plot('High', 'Low', style='o')
plt.plot(dataReg.High, Rez1.fittedvalues)
plt.plot(dataReg.High[:-100], Rez2.fittedvalues)
plt.xlim([0, 4500])
plt.ylim([0, 4500])
plt.show()
```

```python
# Rezime regresije uz pomoc gotovih funkcija summary:
print(Rez1.summary())
print(Rez2.summary())
```

#### Interpretacija rezulata:

#### Df Residuals - broj podataka (Observations) minus broj parametara koje estimiramo
#### Df Model - broj nezavisnih prmenljivih (u ovom slucaju samo jedna: x)
#### R squared - kvadrat indeksa krivolinijske korelacije
#### Adj. R-squared: modifikovan R^2, tako da se bolje uzima u obzir broj parametara (kaznjava koriscenje velikog broja parametara!)
#### F-statistic - vrednost statistike koja se koristi za testiranje hipoteze (H0) da je y nezavisno (u zdruzenom smislu) od svih nezavisnih promenljivih (velika vrednost je u korist odbacivanja hipoteze H0)
####  Prob (F-statistic) - p-vrednost testa, mala vrednost znaci da cemo odbaciti hipotezu o nezavisnosti sa malim nivoom znacajnosti: $\alpha \geq p$
####  Log-Likelihood - Logaritam verodostojnosti dobijenih parametara
#### AIC, BIC - Mere koje govore o kvalitetu modela - manja vrednost znaci bolji model! (bolje kaznjavaju kompleksne modele od R vrednosti)
#### std err - standardna greska (estimacija standardne devijacije) ocena parametara
#### t - vrednost t statistike za testiranje hipoteze (H0) da je y nezavisno od odgovarajuceg x
#### P>|t|  - p -vrednost testa (mala vrednost znaci da odbacujemo hipotezu H0 sa velikom sigurnoscu - malim nivoom znacajnosti)
#### [0.025      0.975] - 95% interval poverenja za parametre
#### Omnibus - test normalnosti reziduala (H0 je hipoteza da jeste normalna raspodela suma!)
#### Prob(Omnibus) - p-vrednost testa normalnosti (jako mala vrednost znaci da odbacujemo hipotezu da je sum normalan!)
####  Skew - ocena nesimetricnosti raspodele reziduala (za normalnu raspodelu (simetricna je)) treba da bude 0
#### Kurtosis - mera "tezine" repova gustine respodele (normalna PDF ima vrednost 3), npr. t-raspodela ima "lakse" repove i time  vecu ovu vrednost
#### Durbin-Watson - mera autokorelacije kod reziduala (prema modelu bi trebalo da su nezavisni, vrednost oko 2 ukazuje na nekorelisanost tj. nezavisnost reziduala)
#### Jarque-Bera (JB) - jos jedan test normalnosti
#### Prob(JB) - p vrednost, jako mala vrednost znaci da odbacujemo hipotezu o normalnosti sa malim nivoom znacajnosti
#### Cond. No. - mera korelisanosti ulaznih promenljivih

Vidimo ogromnu razliku u R^2 indeksu (koristi se za proveru saglasnosti modela sa podacima) u ova dva slucaja

```python
# koristeci popularnu biblioteku za masinsko ucenje: sklearn, odredicemo linearnu regresiju
from sklearn.linear_model import LinearRegression

# ovde je vazno da se radi sa matricama! :

dataM = np.matrix( dataReg )

# sada kreiramo regresione objekte i fitujemo podatke:

cln = LinearRegression()  # izbaceno 100 tacaka
org = LinearRegression()   # originalni skup

X, Y = dataM[:,1], dataM[:,0]
cln.fit( Y[:-100], X[:-100] )  # fitujemo model
org.fit( Y, X )   # sa svim podacima

print(cln.score( X[:-100], Y[:-100] ) )  # .score je R^2 indeks
print(org.score( X, Y ))
    
plt.plot( dataReg.High[:-100], dataReg.Low[:-100], 'bo') 

# Ono cemu najcesce sluzi (obuceni) model je da se vrsi posle predikcija za y u novim tackama X (skup za testiranje - test data):

# generisemo nove tacke:

test = np.c_[np.arange(150, 4400, 1)]  # mora da bude vektor kolona! np._c_ pravi vektor kolonu

plt.plot( test, cln.predict( test ), 'k' )
plt.plot( test, org.predict( test ), 'k--' )
plt.xlim([0, 4500])
plt.ylim([0, 4500])
plt.show()
```

```python
# Ispod je prikaz racunanja Visestruke Regresije
# Racunamo zavisnost za vise od 2 promenljive, tacnije zavisnost jedne zavise promenljive od vise nezavismih
from statsmodels.formula.api import ols

data_visestruka = data_por.loc[:, ["High", "Low", "Volume"]]

modelV = ols("High ~ Low", data_visestruka).fit() # model nezavisan od dnevnog trgovanja
print(modelV.summary())
```

```python
modelV = ols("High ~ Low + Volume", data_visestruka).fit() # zavisnost od dva faktora
print(modelV.summary())
```

```python
modelV2 = ols("High ~ Low * Volume", data_visestruka).fit() # dva faktora + njihova interakcija (proizvod)
print(modelV2.summary())
```

```python
# na kraju radimo estimaciju metodom Nelinearne regresije - polinomske kvadratne
# I dalje radimo linearnu regresiju, samo dodajemo jos jedan ulazni atribut!
# kvadratna regresija:

A=np.array(dataReg['High'])
A=np.array([np.ones(len(A)), A, A**2])
y=np.array(dataReg['Low'])

w2 = np.linalg.lstsq(A.T,y)[0] # nalazenje optimalnih parametara (k,n i e gresku)
print('w2=',w2)
```

```python
# funkcija koja racuna fitovanu vrednost u tacki x, za date parametre w
def f2(x,w):
    return w[0]+w[1]*x+w[2]*x**2


ax=dataReg.plot(kind='scatter',s=20,x='High',y='Low',figsize=[12,8])

# pribavljanje najmanje i najvece vrednosti po x dimenziji, nalazenje opsega izmedju i kreiranje niza
x0,x1=plt.xlim()
opseg=x1-x0
X=np.arange(x0,x1,opseg/100.)

#crtanje tacaka i kvadratne parabolicne funkcije
ax.plot(X,f2(X,w2),'k')
plt.show()
# BOLJI SU REZULTATI ZA LINEARNU REGRESIJU !
```
